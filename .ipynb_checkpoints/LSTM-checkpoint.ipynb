```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data_length = 500
```

```python
x = np.arange(data_length)
y = np.sin(x / 20)
```

```python
#시계열 데이터 플룻 가로, 세로 길이 저장
plt.figure(figsize=(10, 4))
plt.plot(x, y, color='blue', linewidth=2)
plt.xlabel('X')
plt.ylabel('Y = sin(X)')
plt.grid(True)
plt.show()
```

```python
time_steps = 10

X = []
Y = []

for i in range(data_length - time_steps):
    X.append(y[i:i + time_steps])
    Y.append(y[i + time_steps])

X = np.array(X)
Y = np.array(Y)

X = X.reshape(X.shape[0], X.shape[1], 1)
X.shape
```
