### 2D 데이터 생성

```python
np.random.seed(42)
```

2차원 정규분포에서 랜덤 숫자 생성

Loc = 평균, scale =표준편차, size = 결과 배열의 shape

```python
g1 = np.random.normal(loc=(2, 2), scale = 0.5, size=(100, 2))
```

```python
g2 = np.random.normal(loc=(2, 3), scale = 1, size=(100, 2))
```

```python
g3 = np.random.normal(loc=(4, 3), scale=1, size=(100, 2))
```

```python
data = np.vstack((g1, g2, g3))
```

```python
groups = np.repeat([0, 1, 2], 100)
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
```

```python
plt.scatter(data[:, 0], data[:, 1], s=20)
```

```python
plt.legend()
```

```python
plt.title("2D mixed data not identified groups")
```

```python
plt.xlabel('X1')
```

```python
plt.ylabel('X2')
```

```python
plt.subplot(1, 2, 2)
```

```python
for group in np.unique(groups):
    plt.scatter(data[groups == group, 0], data[groups == group, 1],
                 label=f'Group {group}', s=20)
plt.legend()
plt.title("2D data generated by three groups")
plt.xlabel('X1')
plt.ylabel('X2')
```

```python
plt.tight_layout()
```

```python
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
kmeans.fit(data)
```

```python
labels = kmeans.labels_
```

```python
centroids = kmeans.cluster_centers_
```

```python
centroids = kmeans.cluster_centers_
```

```python
centroids
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap = 'viridis', marker='o', s=20)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap = 'viridis', marker='o', s=20)
for i, (x, y) in enumerate(centroids):
    plt.text(x, y, f'Cluster {i}', color='red', fontsize=9,
             ha='center', va='center',
             bbox=dict(facecolor='white', alpha=1, edgecolor='none', boxstyle='round'))
plt.title('K-means Clustering with Centroid Lables')
plt.xlabel('X1')
plt.ylabel('X2')
```

```python
plt.show()
```

```python
import numpy as np
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
X, y = make_moons(n_samples=300, noise=0.1, random_state=20)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=10)
plt.title("Moon data")
plt.show()
```

### K_Means 클러스터링

```python
kmeans = KMeans(n_clusters=2, random_state=10)
labels = kmeans.fit_predict(X)
# 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=10)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', marker='x')
plt.title("K-Means Clustering")
plt.show()
```
